```python
import random
import string

```

```python
#create a function to return a random solution
def generate_random_solution(answer):
    length = len(answer)
    return [random.choice(string.printable) for _ in range (length)]
print(generate_random_solution("Hill One"))
```

```python
#evaluateing a solution for  for the solution s
def evaluate(solution, answer):
    target = list(answer)
    diff =0
    for i  in range(len(target)):
        s=solution[i]
        t=target[i]
        diff += abs(ord(s) - ord(t))
        return diff
print(evaluate(['j', 's', '\\', '3', '%', 'F', '8', 'Z'], "Hill One"))
```

```python
#TO mutate means to change in a small way.  Here we change random letters in our string to arrive at the best solution
def mutate_solution(solution):
    index = random.randint(0, len(solution)-1)
    solution[index] = random.choice(string.printable)
solution = ['j', 's', '\\', '3', '%', 'F', '8', 'Z']
mutate_solution(solution)
print(solution)
```

```python
#Tying together all functions to achieve hill climbing alogorithm
answer = "Hi"
best = generate_random_solution(answer)
best_score = evaluate(best, answer)
while True:
    print('Best score so far' , best_score, 'solution', "".join(best))
    if best_score ==0:
        break
    new_solution = list(best)
    mutate_solution(new_solution)
    
    score = evaluate(new_solution, answer)
    if evaluate(new_solution, answer) < best_score:
        best =new_solution
        best_score = score
```
